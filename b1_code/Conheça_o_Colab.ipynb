# formula da aceleraçao

```
def calculavelMedia(deltas, deltaT):
  velMedia = (deltas/deltaT)
  return velMedia
```

### teste serviço nuvem  aceleraçao media




```
print("Vm=", calculavelMedia(4, 5, 5, 12))
```

### formulas do movimento uniforme (MU)

```
def calculaMU(s, so, v, t):
  s = (so + v*t)
  return s
```

### teste serviços nuvem movimento uniforme

```
print("s=", calculaMU(2, 4, 5, 12))
```
